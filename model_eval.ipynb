```python
import pandas as pd

df = pd.read_csv('perplexity.csv')
df["perplexity ;Dropout 0.0;Dropout 0.2;Dropout 0.4;Dropout 0.6;Dropout 0.8"]
import pandas as pd

df = pd.read_csv('perplexity.csv')
df[['perplexity', 'Dropout 0.0', 'Dropout 0.2', 'Dropout 0.4', 'Dropout 0.6', 'Dropout 0.8']] = df["perplexity ;Dropout 0.0;Dropout 0.2;Dropout 0.4;Dropout 0.6;Dropout 0.8"].str.split(';', expand=True)
df.drop(columns=["perplexity ;Dropout 0.0;Dropout 0.2;Dropout 0.4;Dropout 0.6;Dropout 0.8"], inplace=True)
df
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Adjusting figure size
plt.figure(figsize=(25, 25))

# Convert dropout columns to numeric data types
df[['Dropout 0.0', 'Dropout 0.2', 'Dropout 0.4', 'Dropout 0.6', 'Dropout 0.8']] = df[['Dropout 0.0', 'Dropout 0.2', 'Dropout 0.4', 'Dropout 0.6', 'Dropout 0.8']].apply(pd.to_numeric)

# Plotting the changes of the numbers through all epochs
plt.plot(df['perplexity'], df['Dropout 0.0'], label='Dropout 0.0')
plt.plot(df['perplexity'], df['Dropout 0.2'], label='Dropout 0.2')
plt.plot(df['perplexity'], df['Dropout 0.4'], label='Dropout 0.4')
plt.plot(df['perplexity'], df['Dropout 0.6'], label='Dropout 0.6')
plt.plot(df['perplexity'], df['Dropout 0.8'], label='Dropout 0.8')

# Adding labels and title
plt.xlabel('Epoch')
plt.ylabel('Dropout Values')
plt.title('Changes of Dropout Values through Epochs')



# Adding legend
plt.legend()

# Displaying the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Read the CSV file
df2 = pd.read_csv('valid_perplexity.csv')

# Split the columns and drop the original column
df2[['Valid perplexity', 'Dropout 0.0', 'Dropout 0.2', 'Dropout 0.4', 'Dropout 0.6', 'Dropout 0.8']] = df2['Valid  perplexity ;Dropout 0.0;Dropout 0.2;Dropout 0.4;Dropout 0.6;Dropout 0.8'].str.split(';', expand=True)
df2.drop(columns=['Valid  perplexity ;Dropout 0.0;Dropout 0.2;Dropout 0.4;Dropout 0.6;Dropout 0.8'], inplace=True)

# Adjusting figure size
plt.figure(figsize=(25, 25))

# Convert dropout columns to numeric data types
df2[['Dropout 0.0', 'Dropout 0.2', 'Dropout 0.4', 'Dropout 0.6', 'Dropout 0.8']] = df2[['Dropout 0.0', 'Dropout 0.2', 'Dropout 0.4', 'Dropout 0.6', 'Dropout 0.8']].apply(pd.to_numeric)

# Plotting the changes of the numbers through all epochs
plt.plot(df2['Valid perplexity'], df2['Dropout 0.0'], label='Dropout 0.0')
plt.plot(df2['Valid perplexity'], df2['Dropout 0.2'], label='Dropout 0.2')
plt.plot(df2['Valid perplexity'], df2['Dropout 0.4'], label='Dropout 0.4')
plt.plot(df2['Valid perplexity'], df2['Dropout 0.6'], label='Dropout 0.6')
plt.plot(df2['Valid perplexity'], df2['Dropout 0.8'], label='Dropout 0.8')

# Adding labels and title 
plt.xlabel('Epoch')
plt.ylabel('Dropout Values')
plt.title('Changes of Dropout Values through Epochs')

# Adding legend
plt.legend()

# Displaying the plot
plt.show()
```
